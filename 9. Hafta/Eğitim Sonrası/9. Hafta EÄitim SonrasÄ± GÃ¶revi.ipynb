# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
# 1.Problemin Çözümünü Buraya Yazınız
import random
import time

class Kumanda():
    def __init__(self,marka,model):
        self.marka=marka
        self.model=model
        self.ses=50
        self.acik=False
        self.kanallar=["TRT","SHOW","STAR","NTV","CNN"]
    
    def tv_ac(self):
        
            if self.acik:
                print("Telvizyon zaten açık.")
            else:
                print("Televizyon açılıyor...")
                time.sleep(1)
                self.acik=True
                print("Televizyon açıldı.")

    def tv_kapat(self):
        if self.acik:
            print("Televizyon kapatılıyor...")
            time.sleep(1)
            print("Televizyon kapatıldı.")
        else:
            print("Televizyon zaten kapalı.")

    def ses_ayarla(self):
        if self.acik:
            ses=int(input("Ses seviyesi (0-100):"))
            if ses>=0 and ses <=100:
                self.ses=ses
            else:
                print("Geçersiz ses seviyesi.")
        else:
            print("Televizyon kapalı.")
    
    def kanal_ekle(self):
        if self.acik:
            kanal=input("Kanal:")
            self.kanallar.append(kanal)
            print(f"{kanal} kanal listesine eklendi.")
        else:
            print("Teleziyon kapalı.")
    
    def kanal_sayisi(self):
        print(len(self.kanallar), "kanal var.")
    
    def rastgele_kanal(self):
        if self.acik:
            kanal=random.choice(self.kanallar)
            print("Rastgele kanala geçiliyor: ",kanal)
        else:
            print("Televizyon kapalı.")

    def tv_bilgileri(self):
        print("Marka: ",self.marka)
        print("Model: ",self.model)

    def baslat(self):
        while 1:
            islem=input("TV açmak için '1', TV kapatmak için '2', ses ayarlamak için '3',kanal eklemek için'4',kanal sayısını öğrenmek için '5',rastgele kanala geçmek için'6',televizyon bilgilerini görmek için '7',çıkmak için 'q' tuşuna basın.")
            if islem=="1":
                Kumanda.tv_ac(self)
            elif islem=="2":
                Kumanda.tv_kapat(self)
            elif islem=="3":
                Kumanda.ses_ayarla(self)
            elif islem=="4":
                Kumanda.kanal_ekle(self)
            elif islem=="5":
                Kumanda.kanal_sayisi(self)
            elif islem=="6":
                Kumanda.rastgele_kanal(self)
            elif islem=="7":
                Kumanda.tv_bilgileri(self)
            elif islem=="q":
                break
            else:
                print("Yanlış giriş yaptınız.")

kmnd= Kumanda("SONY","SNY34")
kmnd.baslat()
```

# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
# 1.Problemin Çözümünü Buraya Yazınız
class Vehicle():
    def __init__(self,brand,model,year):
        self.brand=brand
        self.model=model
        self.year=year
        print("Vehicle nesnesi oluşturuldu.")

class Bus(Vehicle):
    def __init__(self, brand, model, year, code):
        super().__init__(brand, model, year)
        self.code=code
        print("Bus nesnesi oluşturuldu.")
    def seating_capacity(self,capacity=50):
        print(f"The seating capacity of a bus is {capacity} passengers")

my_bus=Bus("Mercedes Benz","Travego",1999,"TRVG")
my_bus.seating_capacity()
````
